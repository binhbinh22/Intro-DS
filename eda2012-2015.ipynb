```python
import pandas as pd
```

```python
df = pd.read_csv('data-2012-2015.csv')
df
```

```python
# Chọn các cột cần lấy
selected_columns = ['year', 'dengue','Month', 'temperature','rainfall']
df = df[selected_columns]
df
```

```python
# In thông tin cơ bản 
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Thống kê mô tả của dữ liệu
print(df.describe())
```

```python
# Kiểm tra dữ liệu trống
print(df.isnull().sum())
```

```python
df['Date'] = pd.to_datetime(df[['year', 'Month']].assign(DAY=1))

df_subset = df[(df['Date'] >= '2012-01-01') & (df['Date'] <= '2015-12-31')]

plt.figure(figsize=(12, 6))

# Biểu đồ đường cho số ca mắc (dengue)
plt.subplot(3, 1, 1)
sns.lineplot(x='Date', y='dengue', data=df_subset, label='Dengue', marker='o')
plt.title('Số ca mắc theo từng tháng ')
plt.xlabel('Month')
plt.ylabel('Dengue')
plt.legend()

# Biểu đồ đường cho nhiệt độ (temperature)
plt.subplot(3, 1, 2)
sns.lineplot(x='Date', y='temperature', data=df_subset, label='Temperature', marker='o')
plt.title('Nhiệt độ theo từng tháng ')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.legend()

# Biểu đồ đường cho lượng mưa
plt.subplot(3, 1, 3)
sns.lineplot(x='Date', y='rainfall', data=df_subset, label='Rainfall', marker='o')
plt.title('Lượng mưa theo từng tháng ')
plt.xlabel('Month')
plt.ylabel('rainfall')
plt.legend()

plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
df['Year'] = pd.to_datetime(df['year'], format='%Y').dt.year

# Tính tổng số ca mắc theo năm
total_cases_by_year = df.groupby('Year')['dengue'].sum()

# Vẽ biểu đồ cột
plt.figure(figsize=(10, 6))
total_cases_by_year.plot(kind='bar', color='skyblue')
plt.title('Tổng số ca mắc theo năm')
plt.xlabel('Năm')
plt.ylabel('Tổng số ca mắc')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize distribution của 'dengue' và 'temperature'
plt.figure(figsize=(12, 6))

plt.subplot(3, 2, 1)
sns.histplot(df['dengue'], kde=True)
plt.title('Distribution of Dengue ')

plt.subplot(3, 2, 3)
sns.histplot(df['temperature'], kde=True)
plt.title('Distribution of Temperature')

plt.subplot(3, 2, 2)
sns.boxplot(x='dengue', data=df)
plt.title('Boxplot of Dengue ')

plt.subplot(3, 2, 4)
sns.boxplot(x='temperature', data=df)
plt.title('Boxplot of Temperature')

plt.subplot(3, 2, 5)
sns.histplot(df['rainfall'], kde=True)
plt.title('Distribution of Rainfall')

plt.subplot(3, 2, 6)
sns.boxplot(x='rainfall', data=df)
plt.title('Boxplot of Rainfall')

plt.tight_layout()
plt.show()
```

```python
# Mối quan hệ giữa 'dengue' và 'temperature'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temperature', y='dengue', data=df)
plt.title('Scatterplot giữa số ca và nhiệt độ')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='rainfall', y='dengue', data=df)
plt.title('Scatterplot giữa số ca và lượng mưa')
plt.show()
```

```python
#sự biến thiên của số ca mắc theo năm 
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='dengue', data=df, label='Dengue Cases')
plt.title('Sô ca mắc theo năm ')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='temperature', data=df, label='Temperature')
plt.title('Nhiệt độ theo năm')
plt.show()
```

```python
# Tính ma trận tương quan
correlation_matrix = df[['dengue', 'temperature', 'rainfall']].corr()

# In ra ma trận tương quan
print("Correlation Matrix:")
print(correlation_matrix)

# Tạo heatmap cho mối tương quan giữa 'dengue', 'temperature', và 'rainfall'
plt.figure(figsize=(10, 8))
heatmap_data = df[['dengue', 'temperature', 'rainfall']].corr()
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap among Dengue Cases, Temperature, and Rainfall')
plt.show()
```

không có mối quan hệ tuyến tính mạnh giữa số ca mắc sốt xuất huyết và nhiệt độ.

mối quan hệ tuyến tính tương đối với lượng mưa và số ca mắc sốt xuất huyết.
